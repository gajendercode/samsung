```python
from PIL import Image

root_folder = R"C:\Users\aviator\Music\SIDD_Small_sRGB_Only\Data"

# Iterate through subfolders
for ii,subfolder in enumerate(os.listdir(root_folder)):
    subfolder_path = os.path.join(root_folder, subfolder)

    # Check if it's a directory
    if os.path.isdir(subfolder_path):
        # Assume there are always two images in each subfolder
        images = os.listdir(subfolder_path)

        # Filter noisy images (assuming they contain "noisy" in the filename)
        noisy_images = [img for img in images if "NOISY" in img]

        # Process the noisy images as needed
        for noisy_image in noisy_images:
            noisy_image_path = os.path.join(subfolder_path, noisy_image)
            # Your code to work with the noisy image goes here
            Image.open(noisy_image_path).save(str(ii)+".png")
            print(f"Processing {noisy_image_path}")
```
